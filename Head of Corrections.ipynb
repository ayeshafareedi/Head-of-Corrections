```python
import pandas as pd
import seaborn as sns

from IPython.display import Markdown, display
hoc = pd.read_csv('Corrections.csv')
hoc.head()
```

```python
#Setting the Index as State
hoc.set_index(['State'])
```

# Graph 1: Men vs. Women 

This graph below shows more men lead the corrections department than women. Though this is _**expected**_, it's important to note that 72% of our country's corrections departments are led by men while only 28% are ran by women. 


```python
#Graphing Sex of Head of Corrections
import matplotlib.pyplot as plt
ax = sns.countplot(x="Head of Corrections Sex", data=hoc);
ax.set(xlabel='Number of Men vs. Women', ylabel='Count')

print("Number of Men: ", hoc.loc[hoc['Head of Corrections Sex']== 'M', 'Head of Corrections Sex'].count())
print("Number of Women: ", hoc.loc[hoc['Head of Corrections Sex']== 'F', 'Head of Corrections Sex'].count())
      
```

# Graph 2: Changes in Gubernatorial Administration and the HOC Tenure

This graph shows the number of people who have kept their positions after a new Governor was elected, people appointed by the current governor, and the Interim Appointments as well.


```python
''' Data Prep for Visualizing the Trend in Changing Administrations & the HOC Position'''

#Renaming the Column referring to the Appointment of HOC 
hoc = hoc.rename(columns={'Still in position even after Governor who appointed them leaves office?': 'HOC Appointment Desc'})

#Replacing NaN with 'Appointed by current governor'
hoc['HOC Appointment Desc'] = hoc['HOC Appointment Desc'].fillna('Appointed by Current Governor')

#Replacing Yes with 'Kept Position after Admin Change'
hoc['HOC Appointment Desc'] = hoc['HOC Appointment Desc'].replace('Yes', 'Kept Position after Admin Change')


'''Graphing'''
plt.suptitle('')
ax = sns.countplot(x="HOC Appointment Desc", data=hoc);
ax.set(xlabel='Appointment Description', ylabel='Count');
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right");
```

# Graph 2a: A closer look into Gubernatorial Political Parties and HOCs who kept their positions 

Following the previous graph, this graph below looks into the Political Parties after a change in gubernatorial administrations for HOCs who kept their positions. Rep-Rep & Dem-Dem fall under the 'Same Party' while Rep-Dem & Dem-Rep fall under 'Different Parties'

```python
admin_change = hoc[hoc['HOC Appointment Desc'] == 'Kept Position after Admin Change']
admin_change.loc[admin_change["Governor Appointing HOC's Political Affiliation"] == admin_change["Current Governor's Political Party"],'Political Party Change'] = 'Same Party'

admin_change['Political Party Change'] = admin_change['Political Party Change'].fillna('Different Parties')

ax = sns.countplot(x='Political Party Change', data=admin_change)
```

# Graph 3: HOCs Tenure Histograms

Histograms group data into ranges and the height of each bar shows how many data points fall into that range. The data for these histograms is the Tenure of Current HOCs. 

__Note__: The first histogram omits HOCs appointed after 2016 & the second histogram omits HOCs appointed after 2018. This is simply because HOCs appointed after 2016 and 2018 will have been in the position for a relatively small amount of time. 

The first histogram shows most of the tenure lengths fall between 4 - 7 years while the second one shows most tenure lengths fall between 2 - 4 years. There are a couple of current HOCs who have kept their positions for longer than 10 years.

```python
'''Calculating Tenure in Days for Current HOC and Predecessor HOC'''

#Converting the dates from mm/dd/yyyy to 
hoc['Date of Appointment (HOC)'] = pd.to_datetime(hoc['Date of Appointment (HOC)'])
hoc['HOC Predecessor Tenure Start Date'] = pd.to_datetime(hoc['HOC Predecessor Tenure Start Date'])
hoc['HOC Predecessor Tenure End Date'] = pd.to_datetime(hoc['HOC Predecessor Tenure End Date'])

#Calculating the tenures in Days
hoc['Current HOC Tenure (Days)'] = (pd.to_datetime('7/9/2019') - hoc['Date of Appointment (HOC)'])
hoc['Current HOC Tenure (Years)'] = hoc['Current HOC Tenure (Days)'].dt.days/365
hoc['Predecessor HOC Tenure (Days)'] = hoc['HOC Predecessor Tenure End Date'] - hoc['HOC Predecessor Tenure Start Date']
hoc['Predecessor HOC Tenure (Years)'] = hoc['Predecessor HOC Tenure (Days)'].dt.days/365


'''Graphing Histogram to show the distribution of Current HOC Tenure'''
fig, ax = plt.subplots(1,2);
hoc_current_2016 = hoc[hoc['Date of Appointment (HOC)'].dt.year < 2016]
hoc_current_2018 = hoc[hoc['Date of Appointment (HOC)'].dt.year < 2018]
sns.distplot(hoc_current_2016['Current HOC Tenure (Years)'], ax=ax[0]);
sns.distplot(hoc_current_2018['Current HOC Tenure (Years)'], ax=ax[1]);
fig.subplots_adjust(wspace=.5);

##Time HOC stayed in Office + Time Governor stayed in Office
##sns.lmplot(x='Current HOC Tenure (Years)', data=hoc_current_2016)

"""Entering Titles for Histograms"""
plt.suptitle('Distribution of the Current HOCs Tenure in years');
ax[0].set_xlabel('Past Appointments upto 2015')
ax[1].set_xlabel('Past Appointments upto 2017');
```

__Current HOC whose tenure exceeds 10 years__

Printed below are current head of corrections who have held their position for at least 10 years or more.

```python
hoc[hoc['Current HOC Tenure (Years)'] > 10].loc[:, ['State', 'Head of Corrections', 'Current HOC Tenure (Years)']]
```

# Graph 4: HOC Predecessor's Tenure Histogram

This histogram below is for the most recent HOCs predecessor's tenure. It shows that many of them held the position for about 4 - 5 years. However, there are some who have held the position for longer than 15 years which is unexpected. 

```python
"""Graphing the Histogram for the HOC Predecessors"""
ax = sns.distplot(hoc['Predecessor HOC Tenure (Years)'].dropna());
ax.set_title('Distribution of the HOCs Predecessor Tenure in years');
```

__HOC Predecessors whose tenure exceeds 15 years:__

__NOTE:__ Louisiana and North Dakota are two states which show up in __exceeding__ the expected short tenures of current HOCs and predecessors. 

```python
hoc[hoc['Predecessor HOC Tenure (Years)'] > 15].loc[:, ['State', 'Head of Corrections Predecessor', 'Predecessor HOC Tenure (Years)']]
```

# Graph 5: Predecessor HOC Employment Changes

This graph displays all the different reasons behind the HOCs Predecessor's end of employment. The most common reasons being: retired, replaced, resigned. The least common reasons being: committing a crime and being fired. 

Although, the tenure for this position is relatively short, the data shows that most people leave this position retiring instead of being replaced by a governor. 

```python
plt.suptitle("Predecessor HOC Employment Changes")
ax = sns.countplot(x="HOC Predecessor Reason For Leaving", data=hoc);
ax.set(xlabel='Employment Changes', ylabel='Count');
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right");


print("HOC Predecessor who pled guilty to bribery charges: ",hoc.loc[hoc['HOC Predecessor Reason For Leaving'] == 'Pleaded Guilty to Bribery','Head of Corrections Predecessor'].iloc[0])
print("HOC Predecessor who was fired: ",hoc.loc[hoc['HOC Predecessor Reason For Leaving'] == 'Fired','Head of Corrections Predecessor'].iloc[0]);

```
